```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
```

```python
sentences = [
    "I am learning natural language processing",
    "Natural language processing is part of Machine learning",
    "I enjoy machine lerning",
    "I also enjoy deep leaning",
    "Deep learning uses neural networks",
    "Neural networks is powerful for natural language processing"
]
    
```

```python
#Tokenize
tokenized_sentences = [word_tokenize(s.lower()) for s in sentences]
```

```python
print(tokenized_sentences)
```

```python
model = Word2Vec(
    sentences= tokenized_sentences,
    vector_size=100, # embedding dimension
    window=3,    # context window size - maximum distance b/w target and context words
    min_count=1,  # keep all words - ignroe words appearing less than this number
)
```

```python
print("Similarity b/w machine and lerning")
print(model.wv.similarity("machine","learning"))
```

```python
print("Similarity b/w natural and deep")
print(model.wv.similarity("natural","deep"))
```

```python
print("Most similar words to learning:")
print(model.wv.most_similar("learning"))
```

```python
print("Most similar words to learning:")
print(model.wv.most_similar("machine",topn=5))
```

```python
 # Word2Vec is :
 # Shallow neural network
 # only one hidden layer
 # trained using context prediction
 # static embeddings (same vector for every context )
 # "bank"- money / river - this word2vec cant differentiate
```
