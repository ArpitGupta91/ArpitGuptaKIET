```python
# Predict the intent using a sliding window over sentence embeddings
# This will show us how embeddings
```

```python
import gensim.downloader as api
import numpy as np
from nltk.tokenize import word_tokenize
```

```python
print("Loading model...")
model = api.load("glove-wiki-gigaword-100")
print("Model Loaded...")
```

```python
sentences = "Please open google and check wether"
tokens = word_tokenize(sentences.lower())
```

```python
# SLidng window function
def get_windows(tokens,window_size=2):
    windows = []
    for i in range(len(tokens)):
        left= max(0,i-window_size)
        right=min(len(tokens),i + window_size + 1)
        window_words = tokens[left:right]
        windows.append(window_words)
    return windows;
```

```python
windows= get_windows(tokens,2)
print("WIndows:\n",windows)
```

```python
def window_vector(window):
    vectors = []
    for word in window:
        if word in model:
            vectors.append(model[word])
    if(len(vectors)==0):
        return np.zeros(100)
    return np.mean(vectors,axis=0)
```

```python
window_vectors=[window_vector(w) for w in windows]
print(window_vectors[0].shape)
```
