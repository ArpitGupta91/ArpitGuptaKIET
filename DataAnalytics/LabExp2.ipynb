```python
import pandas as pd
data = {
    "Population": [138000, 33100, 14442, 67886],
    "Capital": ["New Delhi", "Washington D.C.", "Beijing", "Paris"]
}

countries_df = pd.DataFrame(
    data,
    index=["India", "United States", "China", "France"]
)

print(countries_df)
```

```python
countries_df.head()
```

```python
countries_df.info()
```

```python
countries_df.loc["India", "Population"]
```

```python
countries_df[countries_df["Population"] > 50000]
```

```python
countries_df["GDP"] = [3.7, 25.5, 17.7, 3.0]  
```

```python
countries_df
```

```python
countries_df.rename(columns={"Population": "Population_Count"}, inplace=True)
countries_df
```

```python
countries_df.drop(columns=["Capital"], inplace=True)
countries_df
```

```python
countries_df.sort_values(by="Population_Count",inplace=True)
countries_df
```

```python
average_population = countries_df["Population_Count"].mean()
print(average_population)
```
